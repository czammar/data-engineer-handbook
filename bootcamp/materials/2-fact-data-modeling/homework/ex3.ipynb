```python
import psycopg2
from datetime import datetime, timedelta
import time

# Configuración de conexión
conn = psycopg2.connect(
    user="postgres",
    password="postgres",
    host="127.0.0.1",
    port="5432",
    database="postgres"
)
cursor = conn.cursor()

# Rango de fechas
start_date = datetime.strptime("2022-12-31", "%Y-%m-%d")
end_date = datetime.strptime("2023-01-30", "%Y-%m-%d")
current_date = start_date

# Script base
query_template = """
INSERT INTO users_browser_cumulated
WITH
devices AS (
    SELECT 
        device_id,
        browser_type,
        ROW_NUMBER() OVER(PARTITION BY device_id, browser_type) AS row_num
    FROM devices d 
),
devices_dedup AS (
    SELECT 
        device_id,
        browser_type
    FROM devices
    WHERE row_num = 1
),
yesterday AS (
    SELECT
        *
    FROM users_browser_cumulated
    WHERE 1=1
        AND date = DATE('{yesterday}')
),
today_device AS (
    SELECT
        user_id::TEXT,
        e.device_id,
        DATE(event_time) AS date_active,
        devices_dedup.browser_type
    FROM events e 
    LEFT JOIN devices_dedup
    ON 1=1
        AND devices_dedup.device_id = e.device_id
    WHERE 1=1
        AND DATE(e.event_time) = DATE('{today}')
        AND e.user_id IS NOT NULL
        AND e.device_id IS NOT NULL
),
today AS (
    SELECT
        user_id,
        browser_type,
        date_active
    FROM today_device device_data
    GROUP BY 1,2,3
),
results AS (
    SELECT
        COALESCE(t.user_id, y.user_id) AS user_id,
        COALESCE(t.browser_type, y.browser_type) AS browser_type,
        CASE
            WHEN y.dates_active IS NULL THEN ARRAY[t.date_active]
            WHEN t.date_active IS NULL THEN y.dates_active
            ELSE ARRAY[t.date_active] || y.dates_active
        END AS dates_active,
        COALESCE(t.date_active, y.date + INTERVAL '1 day') AS date 
    FROM today t
    FULL OUTER JOIN yesterday y
    ON 1=1
        AND t.user_id = y.user_id
        AND t.browser_type = y.browser_type
)
SELECT * FROM results;
"""

# Ejecutar el script para cada día
while current_date <= end_date:
    yesterday = current_date.strftime("%Y-%m-%d")
    today = (current_date + timedelta(days=1)).strftime("%Y-%m-%d")
    query = query_template.format(yesterday=yesterday, today=today)

    try:
        cursor.execute(query)
        conn.commit()
        print(f"Query ejecutada para fecha: {yesterday} -> {today}")
        time.sleep(1)
    except Exception as e:
        print(f"Error ejecutando query para fecha: {yesterday} -> {today}: {e}")
        conn.rollback()
    
    current_date += timedelta(days=1)

# Cerrar conexión
cursor.close()
conn.close()
```
